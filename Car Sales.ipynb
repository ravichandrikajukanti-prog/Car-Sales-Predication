```python
#import the libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#read CSV file from the 'data' subdirectory using a relative path
data = pd.read_csv('./Car Dataset.csv',index_col=0)
```

```python
#dispaly the first 5 rows
data.head()
```

```python
#display the last 5 rows
data.tail()
```

```python
#check the shape of the dataframe
data.shape
```

```python
#explore data
data.describe()
```

```python
data.isnull().sum()
```

```python
# create a python list of feature names
feature_cols=['TV','Radio','Newspaper']
# use the list to select a subset of the original Dataframe
X=data[feature_cols]
#Equivalent command to do this in one line
X=data[['TV','Radio','Newspaper']]
#print the fist 5 rows
X.head()
```

```python
#check the type and shape of X
print(type(X))
print(X.shape)
```

```python
# use the list to select a subset of the original Dataframe
y=data['Sales']
#Equivalent command to do this in one line
y=data.Sales
#print the fist 5 rows
y.head()
```

```python
#check the type and shape of y
print(type(y))
print(y.shape)
```

```python
#splitting X and y into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X, y,random_state=1)
```

```python
#print training and testing
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#linearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
#interpreting model coefficient
print(linreg.intercept_)
print(linreg.coef_)
```

```python
list(zip(feature_cols,linreg.coef_))
```

```python
#making predictions
y_pred = linreg.predict(X_test)
```

```python
#define true and pred response vaues
true =[100,50,30,20]
pred=[90,50,50,30]
```

```python
#evaluation metrics
#mean absolute error(mae)
mae = metrics.mean_absolute_error(y_test,y_pred)
#mean square error(mse)
mse = metrics.mean_squared_error(y_test,y_pred)
#root mean square error(rmse)
rmse = np.sqrt(mse)
```

```python
#print Evaluation matrics
print("mean absolute error(mae):",mae)
print("mean square error(mse):",mse)
print("root mean square error(rmse):",rmse)
```

```python
#custom rms comupatation(same as rmse)
print("computed RMS (same as rmse):",rmse)
```
